### 2-1. 불용어 제거

```python
def removestopwords(word_tokens):  #불용어 제거 함수
    result = [] 

    for w in word_tokens: 
        if w not in no_list: 
            result.append(w) 
    return result
```

```python
no_list = no2 + no3 + no4  #로드한 불용어만 제거 
df['rm_tokenized_all'] = df['tokenized_all'].apply(removestopwords)
```

```python
no_list =  no1  #직접한 지정한 불용어만 제거
df['rm_tokenized_all2'] = df['tokenized_all'].apply(removestopwords)
```

# Embedding

### 1) stopwords 1~3 제거한 토큰 임베딩

- Skip-gram

```python
# Word2Vec(skip-gram) embedding
from gensim.models import Word2Vec
embedding_model = Word2Vec(df['rm_tokenized_all'], size=100, window = 2, min_count=5, workers=4, iter=100, sg=1)
```

```python
print(embedding_model)
```

```python
###########모델 저장
from sklearn.externals import joblib
filename = 'skipgram_ver1_1.sav'
joblib.dump(embedding_model, filename)
```

```python
# Word2Vec(CBOW) embedding - 성능 안좋음
from gensim.models import Word2Vec
embedding_model = Word2Vec(df['rm_tokenized_all'], size=100, window = 2, min_count=5, workers=4, iter=100, sg=0)
```

# **TSNE**

```python
from sklearn.manifold import TSNE
import gensim.models as g
import matplotlib.pyplot as plt
#model = g.Word2Vec.load(model_name) #원하는 임배딩 모델 불러옴

vocab = list(model.wv.vocab)
X = model[vocab]


tsne = TSNE(n_components=2)

# 500개의 단어에 대해서만 시각화
X_tsne = tsne.fit_transform(X[:500,:])
df_tsne = pd.DataFrame(X_tsne, index=vocab[:500], columns=['x', 'y'])
fig = plt.figure()
fig.set_size_inches(40, 20)
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df_tsne['x'], df_tsne['y'])

for word, pos in df_tsne.iterrows():
    ax.annotate(word, pos, fontsize=30)
plt.savefig("word2vec_500")
plt.show()
```

# 1. Classification

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("summary_embedding_cluster.csv")
df.head()
```

```python
df.drop(['Unnamed: 0', 'summary'], axis=1, inplace=True)
```

```python
X = df.drop(['cluster'], axis=1)
y = df['cluster']
```

### import module 

```python
from sklearn.model_selection import cross_validate
from sklearn.externals import joblib 
```

## 1-1. Multilayer Perceptron Classifier

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier()
```

```python
from sklearn.model_selection import cross_validate
mlp_results = cross_validate(mlp, X, y, cv=10)
```

```python
np.mean(mlp_results['test_score'])
```

### 하이퍼파라미터 튜닝

```python
hidden_layer = [30, 50, 70, 90]
```

```python
for i in hidden_layer:
    mlp = MLPClassifier(hidden_layer_sizes=i)
    mlp_results = cross_validate(mlp, X, y, cv=10)
    print(i, "accuracy:", np.mean(mlp_results['test_score']))
```

```python
#hidden layer가 50일때 accuracy가 가장 높음
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=50)
```

```python
mlp.fit(X,y)
```

```python
#모델 저장
file_name = 'mlp_clf.pkl'
joblib.dump(mlp, file_name)
```

```python
#모델 저장
file_name = 'mlp_clf.sav'
joblib.dump(mlp, file_name)
```
